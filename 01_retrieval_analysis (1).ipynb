```python
# MRR across cutoffs line plot
cutoffs = [5, 10, 20, 50]
colors  = {'BM25': '#E53935', 'Dense': '#1E88E5', 'Hybrid': '#43A047', 'ColBERT': '#FB8C00'}

plt.figure(figsize=(7, 4))
for retriever, values in TABLE2.items():
    plt.plot(cutoffs, values, marker='o', label=retriever,
             color=colors[retriever], linewidth=2)
plt.xlabel('Cutoff k')
plt.ylabel('MRR@k')
plt.title('MRR at Multiple Cutoffs — MS MARCO Passage')
plt.xticks(cutoffs)
plt.legend()
plt.tight_layout()
plt.savefig('../outputs/figures/rq1_mrr_cutoffs.png', dpi=150, bbox_inches='tight')
plt.show()
```
